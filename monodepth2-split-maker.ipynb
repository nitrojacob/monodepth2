```python
#Filter the eigen_full/train_files.txt to create eigen_kaggle/train_files.txt
import os
import PIL.Image as pil

def check_image(img_path):
    return pil.open(img_path).convert('RGB')
    

fd = open('/kaggle/working/monodepth2-master/splits/eigen_full/train_files.txt')
prefix = '/kaggle/input/kitti-eigen/kitti'
out = open('/kaggle/working/train_files.txt', 'w')
lrmap={'l':'02', 'r':'03'}
count = 0; accept = 0; reject = 0;
for line in fd:
    group, seq, lr = line.strip('\n').split(' ')
    seq=int(seq)
    f1 = os.path.join(prefix, group, 'image_{}'.format(lrmap[lr]), 'data', '{:010d}.jpg'.format(seq))
    f2 = os.path.join(prefix, group, 'image_{}'.format(lrmap[lr]), 'data', '{:010d}.jpg'.format(seq+1))
    f3 = os.path.join(prefix, group, 'image_{}'.format(lrmap[lr]), 'data', '{:010d}.jpg'.format(seq-1))
    try:
        check_image(f1)
        check_image(f2)
        check_image(f3)
        out.write(line)
        accept += 1
    except:
        reject += 1
    count += 1
print('processed files:', count, 'accepts:', accept, 'rejects:', reject)
fd.close()
out.close()
```
